```python
import time
def timer(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print("timetaken:",end-start)
        return result
    return wrapper
@timer
def proocessdata():
    print("processing data")
    time.sleep(3)
    print("completed")
proocessdata()
```
